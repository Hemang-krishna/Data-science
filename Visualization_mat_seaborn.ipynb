```python
import matplotlib.pyplot as mplt
import numpy as np
import pandas as pd
```

```python
#matplotlib to render plots in the notebook
%matplotlib inline
```

```python
x = np.arange(1,201,2) 
```

```python
#y = np.random.randint(100)
y = np.random.randint(1,10,100)
```

```python
x
```

```python
y
```

```python
mplt.figure(figsize=(15,5))
mplt.axes(facecolor='pink')
mplt.plot(x,y,color='black' ,label='line chart') #line chart
mplt.xlim(0,200)
mplt.ylim(0,10)
#mplt.xtickets(1)
mplt.xlabel('sales',color='blue')
mplt.ylabel('product')
mplt.title('Line Plot')
mplt.suptitle('sales comparision', size=20, y=1.05)
#mplt.savefig('save.png') # to save
mplt.legend()
mplt.show()
```

```python
import pandas as pd
file_path = "C:/Users/admin/Desktop/Salaries.csv"
salaries=pd.read_csv(file_path)
salaries
```

```python
fig, ax = mplt.subplots(nrows=2, ncols=2)
ax[0,0].set_title('plot 1')
ax[0,0].plot(x,y)
ax[0,1].hist(salaries['salary'])
ax[1,0].boxplot(salaries.salary)
ax[1,1].hist(salaries.service)
mplt.tight_layout()
```

```python
fig, ax1= mplt.subplots(nrows=2, ncols=2, figsize=(10,10))

ax1[0,0].plot(x,y,color='green', linestyle='dashed')
ax1[0,0].set_title('line')
ax1[0,1].set_title('piechart')
ax1[0,1].pie(y)
ax1[1,0].set_title('scatter plot')
ax1[1,0].scatter(x,y)
ax1[1,1].set_title('Histogram')
ax1[1,1].hist(y,color='orange')
mplt.show()
```

```python
import seaborn as sn
```

```python
tips = sn.load_dataset('tips')
tips.head()
```

Univariate -> "one variable" data visualization

```python
#strip plot
sn.stripplot(x='sex',y='tip',data=tips,jitter=True)
mplt.title('strip plot')
mplt.xlabel('gender')
mplt.show()
```

```python
sn.pairplot(tips)
#plt.tight_layout()
mplt.show()
```
